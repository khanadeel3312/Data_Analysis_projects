```python
w_l.fillna(w_l.mean(),inplace=True)
```

```python
d_a['ActivityDate']=pd.to_datetime(d_a['ActivityDate'])
Date=d_a['ActivityDate'].dt.date
d_a.drop(columns=['ActivityDate'],inplace=True)

d_a.insert(1,'Date1',Date)
```

```python
s_d['SleepDay']=pd.to_datetime(s_d['SleepDay'])
Date=s_d['SleepDay'].dt.date
s_d.drop(columns=['SleepDay'],inplace =True)
s_d.insert(1,'Date1',Date)
```

```python
w_l['Date']=pd.to_datetime(w_l['Date'])
date=w_l['Date'].dt.date

w_l.drop(columns=['Date'],inplace=True)
w_l.insert(1,'Date1',date)
```

```python
merged1= pd.merge(s_d, d_a, on=['Id', 'Date1'], how='right')
merged=pd.merge(merged1,w_l,on=['Id', 'Date1'], how='right')
merged
```

```python
conditions = [
    (merged['BMI'] >= 30),
    (merged['BMI'] >= 25),
    (merged['BMI'] < 25)
]
choices = ['Obese', 'Healthy', 'Underweight']

merged['BMIRange'] = np.select(conditions, choices, default='Unknown')
```

```python
merged['Date1'] = pd.to_datetime(merged['Date1'])
merged['Weekday']=merged['Date1'].dt.strftime('%A')
merged['Weekday']
```

```python
selected=['Id', 'Date1','TotalSleepRecords', 'TotalMinutesAsleep', 'TotalTimeInBed', 'TotalSteps', 'TotalDistance','VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories', 'WeightKg', 'BMI', 'Weekday']
merged1=merged[selected]
merged1
```

```python
result = (
    merged1.groupby('Weekday')
    .agg({'TotalSteps': 'mean', 'Calories': 'mean'})
    .sort_values(by='Weekday')
    .reset_index()
    .rename(columns={'TotalSteps': 'MeanTotalSteps', 'Calories': 'MeanCalories'})
   
)
result
```

```python
merged1['AwakeTimeInBed']=merged1['TotalTimeInBed']-merged1['TotalMinutesAsleep']
merged1['AwakeTimeInBed']
```

```python
result = (
    merged1.groupby('Weekday')
    .agg({'TotalMinutesAsleep': 'mean', 'TotalTimeInBed': 'mean','TotalTimeInBed':'mean','AwakeTimeInBed':'mean'})
    .reset_index())
    
   
 
result
```

```python
activity=(
    merged.groupby('Id')
    .agg({'VeryActiveMinutes':'mean','FairlyActiveMinutes':'mean','LightlyActiveMinutes':'mean','SedentaryMinutes':'mean'})
    .reset_index())
activity
```
